### This example demonstrates how to use `spdc.py` functions to generate the JSA and other properties of the **double-pass** structure with **aperiodically** poled nonlinear waveguides under the assumption of *linear dispersion*. 

```python
import matplotlib.pyplot as plt
import numpy as np
from NeedALight.spdc import SPulsed_lin, JSA
from custom_poling.core.target import Target
from custom_poling.core.custom_crystal import CustomCrystal
from tqdm import tqdm
```

This section introduces all relevant quantities needed to use the functions and generate the wanted properties.

```python
N = 301  # Number of frequency values
Np = 0.6 # Number of pump photons  0.6 for ~10 NS double pass
vp = 0.1  # pump velocity
l = 1.0  # amplification region length
sig = 1  # pump wave packet spread
a = 1.61 / 1.13  # from symmetric grp vel matching

def symmetric_v(vp, sig, l, a):
    vi = vp / (1 - 2 * a * vp / (l * sig))
    vs = vp / (1 + 2 * a * vp / (l * sig))
    return vs, vi

vs, vi = symmetric_v(vp, sig, l, a)

#Frequency values
wi = -10
wf = 10
w = np.linspace(wi, wf, N)
dw = np.abs(w[1]-w[0])

#Defining gaussian pump pulse
def pump(x, Np, scale=1):
    return np.sqrt(Np)*np.exp(-((x) ** 2) / (2 * (sig*scale) ** 2)) / np.power(np.pi * (sig*scale)**2, 1 / 4)

# Crystal properties
Ndomain = 1000
dz = l / Ndomain
domain_width = dz
number_domains = Ndomain
L = number_domains * domain_width
k0 = np.pi / domain_width

# Numerical integration parameters for generating domain
k_range = 100/L
dk = k_range/401
k_array = np.arange(k0-k_range/2,k0+k_range/2,dk)
```

Here we generate the crystal domain which produces a gaussian phase-matching function. We use the code and follow the examples of [Custom-Poling](https://github.com/abranczyk/custom-poling)

```python
# Create a custom crystal object
custom_crystal_gauss = CustomCrystal(domain_width,number_domains)
domain_middles_gauss = custom_crystal_gauss.domain_middles

#Define and plot the target function
std = 10
height = 0.025
target_pmf_gauss = lambda k:1j*height*np.exp(-(k-k0)**2/(2*std**2))*np.exp(1j * L/2 * k)
target_gauss = Target(target_pmf_gauss,k_array)

# Compute and plot the target amplitude
target_amplitude_gauss = target_gauss.compute_amplitude(k0,domain_middles_gauss)

# Compute and plot the custom domains
custom_domains_gauss = custom_crystal_gauss.compute_domains(target_amplitude_gauss,k0)

#Shifting the domain such that we are at 0 central frequency
if len(custom_domains_gauss)==Ndomain:
    domain = custom_domains_gauss * np.asarray([1, -1] * int(Ndomain / 2)) 
else:
    domain = custom_domains_gauss * np.asarray([1, -1] * int(Ndomain / 2)+[1]) 


#Proper scale factor so that Gaussian PMF and Gaussian Pump have the same width
sc=std/(2*a)
```

For the double pas structure, we generate each pass individually and we do not remove the free-propagating phases. The total propagator is given by the product of both single passes. To mimic the polarization swap, we swap the velocities of the signal and idler for the second pass. Furthermore, to mimic reflecting back into the same crystal, we flip the crystal domain. 

```python
#Initializing Pump
pumpscaled=lambda x: pump(x, Np*np.sqrt(sc), scale=sc) #Scale is modified with poling, defined in the previous cell.

#Pass 1, note that for the double pass we do not remove the free-propagating phases for the JSA
T, _J, _Ns, _K, _M, _Nums, _Numi = SPulsed_lin(vs,vi,vp,pumpscaled,domain,dz,l,w, rmv=False)
#Pass 2, assuming pump is unmodified between passes.
#We signal and idler velocities (to mimic polarization swap). 
#We flip domain to mimic passing through in reverse.
T2, _J, _Ns, _K, _M, _Nums, _Numi = SPulsed_lin(vi,vs,vp,pumpscaled,np.flip(domain),dz,l,w,rmv=False)

#Total propagator
T3 = T2@T

#Using JSA function to obtain results desired results
J, Ns, K, M, _Nums, _Numi = JSA(T3,dw)



fig, ax = plt.subplots()

im = ax.imshow(np.abs(J), cmap="jet", origin="lower", extent=[w[0],w[N-1],w[0],w[N-1]])
ax.set_xlabel(r"$(\omega_{s}-\bar{\omega}_{s})/\sigma$")
ax.set_ylabel(r"$(\omega_{i}-\bar{\omega}_{i})/\sigma$")
ax.set_title(r"$|J(\omega_{s}, \omega_{i})|    \langle Ns \rangle = %1.2e $" %Ns)
plt.colorbar(im, orientation="vertical")
plt.show()
print(K)
```

To obtain the only relevant Schmidt mode (high spectral purity), we use the fact that for the symmetric-group-velocity matching regime, the ouput signal mode of the double-pass is equivalent to the flip of the output signal mode of the single pass (c.f. [AVS Quantum Sci. 6, 021402 (2024)](https://doi.org/10.1116/5.0203013)). The output idler mode is of the double-pass is equivalent to the output signal mode of the single pass.

```python
#Generating modes for single pass without free-propagating phases
_T, _J, _Ns, _K, _M, Nums, _Numi = SPulsed_lin(vs,vi,vp,pumpscaled,domain,dz,l,w)

mark=1
val, u = np.linalg.eigh(Nums)


fig, ax1 = plt.subplots()
ax1.plot(w, np.flip(np.real(np.exp(-1j*np.angle(u[(N-1)//2,-mark]))*u[:,-mark])), label='Re') #Here we multiply by a phase factor such that the imaginary part is 0 at the origin.
ax1.plot(w, np.flip(np.imag(np.exp(-1j*np.angle(u[(N-1)//2,-mark]))*u[:,-mark])), label='Im')
ax1.set_xlabel(r"$\omega$")
ax1.set_ylabel(r"$Amplitude$")
ax1.set_title(r"$\langle Ns \rangle= %1.2e$" %Ns)
ax1.legend()
plt.show()
```

Here we calculate the fidelity/overlap between a fixed low-gain Schmidt mode and another one with variable gain. We loop over different pump strengths.

```python
#Overlaps for dual passes with different Strengths (each pass has same strength)
npi = 0.00006
npf = 0.6
nstep = 10
num_pump = np.linspace(npi,npf,nstep)
nsv = np.zeros_like(num_pump)
K = np.zeros_like(num_pump)
Temp_mode = np.zeros([len(w), len(num_pump)],dtype= np.complex128)
overlap = np.zeros(len(num_pump))


for i in tqdm(range(np.size(num_pump))):
    pump_temp = lambda x: pump(x,num_pump[i]*np.sqrt(sc),scale=sc)
    #Pass 1   
    T1T, _J, _Ns, _K, _M, _Nums, _Numi = SPulsed_lin(vs,vi,vp,pump_temp,domain,dz,l,w, rmv=False)
    #Pass 2
    T2T, _J, _Ns, _K, _M, _Nums, _Numi = SPulsed_lin(vi,vs,vp,pump_temp,np.flip(domain),dz,l,w,rmv=False)

    J, NsT, Schmidt, _M, NumsT, _Numi = JSA(T2T@T1T,dw)

    K[i] = Schmidt
    nsv[i] = NsT

    u, v = np.linalg.eigh(NumsT)
    Temp_mode[:,i] = v[:,-1]
overlapmat = np.abs(np.conj(Temp_mode).T @ Temp_mode) ** 2
overlap[:] = overlapmat[0,:]
```

```python
#Plots the overlaps and Schmidt numbers as a function of Ns
fig, axs = plt.subplots(1, 2, figsize=(10, 4))
axs[0].plot(nsv, overlap)
axs[0].set_xlabel(r"$\langle N_{s} \rangle$")
axs[0].set_ylabel(r"$\left| \langle  \psi_{N_{s}}  | \psi_{0}   \rangle \right|^{2}$")
axs[0].set_title("Overlap of low and high gain temporal modes")

axs[1].plot(nsv, K)
axs[1].set_xlabel(r"$\langle N_{s} \rangle$")
axs[1].set_ylabel("K")
axs[1].set_title("Schmidt Number")

fig.tight_layout()
plt.show()
```
